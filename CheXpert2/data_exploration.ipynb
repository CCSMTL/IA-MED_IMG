```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from skimage import feature, filters
from PIL import Image
```

```python
files = pd.read_csv("/mnt/c/Users/bejo2361/PycharmProjects/IA-MED_IMG/data/ChexPert.csv")
```

```python
idx=0

img = cv2.imread(f"/mnt/c/Users/bejo2361/PycharmProjects/IA-MED_IMG{files.at[idx,'Path']}")
print(img.shape)
plt.imshow(img)
plt.show()
```

```python

#plt.imshow(img)
#plt.show()
gray_img =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


ret, thresh = cv2.threshold(gray_img, 50, 255, 0)
contours= cv2.findContours(thresh,cv2.RETR_LIST, cv2.CHAIN_APPROX_SIMPLE)[0]
bool_mask=img.copy()
cv2.drawContours(bool_mask, contours, -1, (255,255,255), -1)



plt.imshow(bool_mask)
plt.show()
```

```python

gray_img =  255-cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img,cmap="gray")
plt.show()
img_gray=cv2.equalizeHist(gray_img)
plt.imshow(img_gray,cmap='gray')
plt.show()
```

```python

hpf_kernel = np.full((3, 3), -1)
hpf_kernel[1,1] = 9
img = cv2.filter2D(img, -1, hpf_kernel)
plt.imshow(img,cmap='gray')
plt.show()
```

```python

sobel = filters.sobel(img)
plt.imshow(sobel,cmap="gray")
plt.show()
```
